```python
# importing the dependencies

import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
```

```python
#selection_feature = movies_data[['Movie_Genre','Movie_Keywords','Movie_Tagline', 'Movie_Cast','Movie_Director']].fillna('')
```

```python
selection_feature = ['Movie_Genre','Movie_Keywords','Movie_Tagline', 'Movie_Cast','Movie_Director']
```

```python
for feature in selection_feature:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_feature = movies_data['Movie_Genre'] +' '+movies_data['Movie_Keywords'] +' '+movies_data['Movie_Tagline'] +' '+movies_data['Movie_Cast'] +' '+movies_data['Movie_Director']
```

```python
print(combined_feature)
```

```python
vectorizer = TfidfVectorizer()
```

```python

feature_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python

similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
movie_name = input('enter you favorite movie name')
```

```python
list_of_all_titles = movies_data['Movie_Title'].tolist()
```

```python
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
```

```python
find_close_match
```

```python
close_match = find_close_match[0]
```

```python
index_of_the_movie = movies_data[movies_data.Movie_Title == close_match]['Movie_ID'].values[0]
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
```

```python
print('movies suggested for you\n')
```

```python
i =1
for movie in sorted_similar_movies:
  index = movie[0]
  title_form_index = movies_data[movies_data.Movie_ID == index]['Movie_Title'].values
  if (i<11):
    print(i,'.',title_form_index)
    i+=1
```
